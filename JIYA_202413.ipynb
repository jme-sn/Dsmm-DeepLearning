```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
dfForest = pd.read_csv("Algerian_forest_fires_cleaned.csv",)
dfForest.head()
```

Data Preparation

```python
dfForest.info()
```

```python
dfForest.describe()
```

```python
dfForest.duplicated().sum()
```

```python
dfForest.plot(
    kind='box', 
    subplots=True, 
    sharey=False, 
    figsize=(20, 7)
)
plt.subplots_adjust(wspace=1) 
plt.show()
```

```python
dfForest.columns
```

```python
#Removing outliers for every other columns other than RH
for column in ['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Region']:
    Q1 = dfForest[column].quantile(0.25)
    Q3 = dfForest[column].quantile(0.75)
    IQR = Q3 - Q1
    lowerBound = Q1 - 1.5 * IQR
    upperBound = Q3 + 1.5 * IQR
    dfForest = dfForest[(dfForest[column] >= lowerBound) & (dfForest[column] <= upperBound)]
```

```python
dfForest.shape
```

```python
dfForest.plot(
    kind='box', 
    subplots=True, 
    sharey=False, 
    figsize=(20, 7)
)
plt.subplots_adjust(wspace=1) 
plt.show()
```

```python
corr_matrix = dfForest.corr()
plt.figure(figsize=(12, 7))  
sns.heatmap(corr_matrix, annot=True,  cmap='coolwarm')
plt.show()
```

```python
#Taking all columns for training the first model
X = dfForest[['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'Region']]
Y = dfForest['FWI']
```

```python
trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
sr = StandardScaler()
trainXsc = sr.fit_transform(trainX)
testXsc = sr.transform(testX)
```

Model Training 1

```python
#Linear Regression
lrModel1 = LinearRegression()
```

```python
lrModel1.fit(trainXsc, trainY)
```

```python
lrModel1.coef_, lrModel1.intercept_
```

```python
yPred = lrModel1.predict(testXsc)
```

```python
#Metrics for Model 1
mse = mean_squared_error(testY,yPred)
mae = mean_absolute_error(testY,yPred)
rsq = r2_score(testY,yPred)
print("LR Model 1 MSE : ", mse)
print("LR Model 1 MAE : ", mae)
print("LR Model 1 R2 : ", rsq)
```

Model Training 2

```python
#After considering correlation and Coefficients from model 1 removed some columns
X = dfForest[['month', 'Temperature', 'RH', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI']]
trainX, testX, trainY, testY = train_test_split(X, Y, test_size=0.2, random_state=42)
sr = StandardScaler()
trainXsc = sr.fit_transform(trainX)
testXsc = sr.transform(testX)
```

```python
lrModel2 = LinearRegression()
lrModel2.fit(trainXsc, trainY)
lrModel2.coef_, lrModel2.intercept_
```

```python
yPred = lrModel2.predict(testXsc)
mse = mean_squared_error(testY,yPred)
mae = mean_absolute_error(testY,yPred)
rsq = r2_score(testY,yPred)
print("LR Model 2 MSE : ", mse)
print("LR Model 2 MAE : ", mae)
print("LR Model 2 R2 : ", rsq)
```

Pickling The Model

```python
import pickle
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(lrModel2, file)
with open('scaler_model.pkl', 'wb') as file:
    pickle.dump(sr, file)
```

Reading Pickled Files

```python
with open('linear_regression_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
    
with open('scaler_model.pkl', 'rb') as file2:
    loaded_scaler = pickle.load(file2)
```
